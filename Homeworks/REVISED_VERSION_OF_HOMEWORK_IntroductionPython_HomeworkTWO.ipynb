# **First Version:**

```python
default_username = "user12"
default_password = "1234"
```

```python
while True:
  username = input("Enter username: ")
  password = input("Enter password: ")

  if (default_username == username and default_password == password):
    print("Login successful")
    break
  else:
    print("Incorrect password, try again...")
```

# **Second Version:**
> This version uses a dictionary !!!

```python
account_dict = {
  "Default Username" : "user12",
  "Default Password" : "1234"
}
```

```python
while True:
  username = input("Enter username: ")
  password = input("Enter password: ")

  if (account_dict["Default Username"] == username and account_dict["Default Password"] == password):
    print("Login successful")
    break
  else:
    print("Incorrect password, try again...")
```
