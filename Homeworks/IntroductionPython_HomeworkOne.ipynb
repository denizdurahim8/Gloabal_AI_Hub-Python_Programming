```python
# List One {Odd values from 0 to 100}:
list_one = [a for a in range(101) if a % 2 == 1]
list_one
```

```python
# List One {Even values from 0 to 100}:
list_two = [b for b in range(101) if b % 2 == 0]
list_two
```

```python
# Merging the lists:
list_three = list_one + list_two
list_three.sort()
list_three
```

```python
# Final List with doubled values:
final_list = [number*2 for number in list_three]
final_list
```

```python
# Printing the data types of all the value inside the list:

for num in range(len(final_list)):
  print("Type of element {} which has the value of {} is {}".format((num+1), (final_list[num]), (type(final_list[num]))))
```
