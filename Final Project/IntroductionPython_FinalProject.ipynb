# **Required Functions:**

```python
# Method for creating questions

def create_question(question, option_a, option_b, option_c, option_d, correct_answer):
  return_dictionary = {
    "Question" : question,
    "Option A" : option_a,
    "Option B" : option_b,
    "Option C" : option_c,
    "Option D" : option_d,
    "Correct Answer" : correct_answer
  }

  return return_dictionary
```

```python
# Method for printing the questions

def print_question(question_dictionary):
  print(question_dictionary["Question"])
  print("A) " + question_dictionary["Option A"])
  print("B) " + question_dictionary["Option B"])
  print("C) " + question_dictionary["Option C"])
  print("D) " + question_dictionary["Option D"])
```
